```python
import pandas as pd
import numpy as np
```

```python
colunas = ['NU_INSCRICAO',
           'SG_UF_PROVA', 
           'TP_PRESENCA_CN', 
           'TP_PRESENCA_CH', 
           'TP_PRESENCA_LC', 
           'TP_PRESENCA_MT', 
           'NU_NOTA_CN', 
           'NU_NOTA_CH', 
           'NU_NOTA_LC', 
           'NU_NOTA_MT', 
           'NU_NOTA_REDACAO']
```

```python
file = 'MICRODADOS_ENEM_2021.csv'
ano = '2021'
```

```python
def analisar_nan(coluna):
    df = dados.loc[dados[coluna].isna(), coluna]
    return(len(df))
```

```python
dados = pd.read_csv(file, usecols = colunas, sep = ';', encoding = 'ISO-8859-1')
```

```python
dados.head()
```

```python
cont = 0
for coluna in colunas:
    nan = analisar_nan(coluna)
    if nan != 0:
        cont += 1
        print(f'Coluna {coluna} tem {(nan):,.0f} valores vazios {(nan / len(dados)):.2%}.\n')
print(f'Foram encontradas {cont} colunas com valores vazios.')
```

```python
dados.loc[dados['TP_PRESENCA_CN'].isna()]
```

```python
dados_2021 = dados[colunas]
```

```python
dados_2021.head()
```

```python
dados_2021.info()
```

```python
colunas_presenca = ['TP_PRESENCA_CN', 'TP_PRESENCA_CH', 'TP_PRESENCA_LC', 'TP_PRESENCA_MT']
for coluna in colunas_presenca:
    dados_2021[coluna].fillna(0, inplace = True)
    dados_2021[coluna] = dados_2021[coluna].astype(np.int64)
```

```python
colunas_notas = ['NU_NOTA_CN', 'NU_NOTA_CH', 'NU_NOTA_LC', 'NU_NOTA_MT', 'NU_NOTA_REDACAO']
for coluna in colunas_notas:
    dados_2021[coluna] = dados_2021[coluna].fillna(0)
```

```python
dados_2021.info()
```

```python
dados_2021['MEDIA'] = 0.0
```

```python
import time

tempo_inicial = time.time()

for i in range(len(dados_2021)):
    soma_notas = 0
    for coluna in colunas_notas:
        soma_notas = soma_notas + dados_2021[coluna][i]
    media_linha = soma_notas/len(colunas_notas)
    dados_2021['MEDIA'][i] = media_linha
    if i%1000 == 0:
        print(f'{i} - {(i/len(dados_2021)):.2%}')

tempo_final = time.time()
duracao = tempo_final - tempo_inicial
print(f'O processamento durou {(duracao)} segundos.')
```

```python
estados = ['AC', 'AM', 'AP', 'PA', 'RO', 'RR', 'TO',
          'AL', 'BA', 'CE', 'MA', 'PI', 'PE', 'PB', 'RN', 'SE', 
          'RS', 'SC', 'PR',
          'GO', 'MT', 'MS', 'DF',
          'SP', 'RJ', 'MG', 'ES']
estados.sort()
```

```python
dicionario_medias = {}
for estado in estados:
    media_estado = dados_2021.loc[dados_2021['SG_UF_PROVA'] == estado, 'MEDIA'].mean()
    dicionario_medias[estado] = media_estado
    print(estado, media_estado)    
```

```python
dicionario_medias
```

```python
df = pd.DataFrame.from_dict(dicionario_medias, orient='index', columns = ['Média ' + ano])
df
```

```python
# Ausências nos dois dias:
ausencias_1_e_2_dias = dados_2021.loc[
    (dados_2021['TP_PRESENCA_CN'] == 0) & 
    (dados_2021['TP_PRESENCA_MT'] == 0) & 
    (dados_2021['TP_PRESENCA_LC'] == 0) & 
    (dados_2021['TP_PRESENCA_CH'] == 0)]
print(f'Faltas nos dois dias: {(len(ausencias_1_e_2_dias)):,.0f}.')
```

```python
dicionario_ausentes = {}

for estado in estados:
    inscritos_estado = len(dados_2021.loc[dados_2021['SG_UF_PROVA'] == estado])    
    ausencias_1_e_2_dia_estado = len(dados_2021.loc[
        (dados_2021['SG_UF_PROVA'] == estado) & 
        (dados_2021['TP_PRESENCA_LC'] == 0) & 
        (dados_2021['TP_PRESENCA_CH'] == 0) & 
        (dados_2021['TP_PRESENCA_CN'] == 0) & 
        (dados_2021['TP_PRESENCA_MT'] == 0)])
    print(f'Estado: {estado}')
    print(f'Ausentes 1º e 2º dias: {(ausencias_1_e_2_dia_estado):,.0f} = {(ausencias_1_e_2_dia_estado / inscritos_estado):.2%}.')
    print('=======================================')
    dicionario_ausentes[estado] = 100*ausencias_1_e_2_dia_estado / inscritos_estado
```

```python
df2 = pd.DataFrame.from_dict(dicionario_ausentes, orient='index', columns = ['Faltas primeiro e segundo dias (%)'])
df2
```
