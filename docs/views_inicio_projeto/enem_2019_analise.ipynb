### Arquivos dispon√≠veis em <https://www.gov.br/inep/pt-br/acesso-a-informacao/dados-abertos/microdados/enem>

```python
import pandas as pd
```

```python
url_or_file = 'MICRODADOS_ENEM_2019.csv'
```

```python
colunas = ['TP_PRESENCA_CN', 
           'TP_PRESENCA_CH', 
           'TP_PRESENCA_LC', 
           'TP_PRESENCA_MT', 
           'NU_NOTA_CN', 
           'NU_NOTA_CH',
           'NU_NOTA_LC',
           'NU_NOTA_MT',
           'NU_NOTA_REDACAO']
```

```python
dados = pd.read_csv(url_or_file, usecols = colunas, sep = ';', encoding = 'ISO-8859-1')
```

```python
dados.info()
```

```python
dados.head(10)
```

```python
colunas_presenca = ['NU_NOTA_CN', 'NU_NOTA_CH', 'NU_NOTA_LC', 'NU_NOTA_MT', 'NU_NOTA_REDACAO']
dados_presentes = dados[colunas_presenca]
```

```python
dados_presentes.head(10)
```

```python
dados_presentes = dados_presentes.dropna(how = 'all', axis = 0)
```

```python
dados_presentes.head(10)
```

```python
print(f'Total de inscritos no ENEM 2019: {(len(dados)):,.0f}.')
print(f'Total de inscritos que registraram alguma presen√ßa: {(len(dados_presentes)):,.0f} ({(len(dados_presentes) / len(dados)):.2%}).')
print(f'Total de inscritos ausentes nos dois dias de provas: {(len(dados) - len(dados_presentes)):,.0f} ({((len(dados) - len(dados_presentes)) / len(dados)):.2%}).')
```
