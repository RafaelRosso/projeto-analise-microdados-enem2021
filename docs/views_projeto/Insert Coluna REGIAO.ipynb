```python
# Importar o pandas
import pandas as pd
```

```python
# Acesso ao arquivo CSV
url_or_file = 'Dados_Enem_2021.csv'
```

```python
# Tratar o arquivo
dados = pd.read_csv(url_or_file, sep = ',', encoding = 'utf-8')
```

```python
dados.head()
```

### 1 - Criar variáveis de Regiões

```python
# Criar variáveis de Regiões "Lista UF = Regiões"
Sul = ['PR', 'SC', 'RS']
Norte = ['AM', 'PA', 'RR', 'AC','RO','AP', 'TO']
Nordeste = ['BA', 'CE','MA', 'PB', 'PE', 'RN', 'SE', 'AL', 'PI']
Sudeste = ['SP', 'RJ', 'MG','ES']
CentroOeste = ['MT', 'MS', 'GO', 'DF']
```

### 2 - Inserir coluna REGIAO

```python
# Inserir coluna REGIAO
dados.insert(10, 'REGIAO', 'NaN')
dados.head()
```

```python
# Ler colunas
display(dados.info())
```

### 3 - Popular coluna 'REGIAO'

```python
# Popular coluna 'REGIAO' através da coluna 'SG_UF_PROVA' 
dados['REGIAO'] = dados['SG_UF_PROVA'].apply(lambda x: 'Sul' if x in Sul else 'Norte' if x in Norte else 'Nordeste' if x in Nordeste else 'Sudeste' if x in Sudeste else 'CentroOeste')
dados.head()
```
